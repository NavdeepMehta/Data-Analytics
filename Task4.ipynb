```python
# Remove columns which has more than 50% missing value
lst=[]
for i in df.columns:
    x=df[i]
    if (x.isnull().sum()/len(df)*100)>50:
        lst.append(i)
    else:
        pass
df.drop(lst,axis =1,inplace = True )
df.head()
```

```python
#rename the columns which aren't well labelled 

df.rename(columns={"iyear" : "Year", "imonth" : "Month", "iday" : "Day", "country_txt" : "Country","region_txt" : "Region",
                   "provstate" : "State", "summary" : "Summary", "attacktype1_txt" : "AttackType", 
                   "targtype1_txt" : "TargetType", "gname" : "Group", "weaptype1_txt" : "WeaponType", 
                   "nkill" : "Kill", "nwound" : "Wound",}, inplace=True)
```

```python
#keeping only the columns which can be responsible for terrorist attack 
df = df[["Year", "Month", "Day", "AttackType", "TargetType", "Country", "Region", "Summary", "State",  "WeaponType", "Kill", "Wound", "Group"]]
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df['Kill'].fillna(df['Kill'].mode()[0],inplace=True)
df['Wound'].fillna(df['Wound'].mode()[0],inplace=True)
df['State'].fillna('Unknown',axis=0,inplace=True)
df['Summary'].fillna('Unknown',axis=0,inplace=True)
```

```python
df = df.astype({'Kill':'int64','Wound':'int64'})
```

```python
temp =pd.DataFrame(index=df.columns)
temp['Data_Types'] = df.dtypes
temp['Null_Count'] = df.isnull().sum()
temp['Unique_Counts'] =df.nunique()

temp
```

```python
for i in range(len(df)):
    if (df['Day'][i]) == 0:
        df['Day'][i] =1
    else :
        pass
for i in range(len(df)):
    if df['Month'][i] == 0:
        df['Month'][i] =1
    else :
        pass    
```
